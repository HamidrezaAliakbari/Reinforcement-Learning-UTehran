# Problem #3 - Confidence Interval

**A) Construct a 95% confidence interval on the sample means of my_kappa distribution. Your result, obviously, consists of two numbers: a lower bound and an upper bound of the calculated interval.**

```python
mu = drawn_samples_total_mean
sigma = drawn_samples_total_stdv
scale = 90
samplesize = 200;
interval = 0.975
b = stats.norm.ppf(interval)
fig = plt.figure(figsize=(15,7))
fig.patch.set_facecolor('yellow')
fig.patch.set_alpha(0.47)
ax = fig.add_subplot(111)
ax.hist(drawn_samples_mean, 70, density=False, color='orange', edgecolor = 'black')
ax.plot([mu,mu],[0, 1*scale], 'k-', lw=4, color='white')
ax.plot([mu-(b*sigma),mu-(b*sigma)],[0, .5*scale], 'k-', lw=3, color='tomato')
ax.plot([mu+(b*sigma),mu+(b*sigma)],[0, .5*scale], 'k-', lw=3, color='tomato')
ax.patch.set_facecolor('red')
ax.patch.set_alpha(0.2)
plt.title("Distribution of 2000-samples-of-size-100's mean from my_Kappa")
plt.xlabel('Samples magnitude deviation')
plt.ylabel('Density of samples in each bin')
plt.show()
print('Lower bound : ', mu-(b*sigma))
print('Upper bound : ', mu+(b*sigma))
```

**B) What does this interval mean?**

> refers to the probability of 95% that  parameter will fall between lower bound and upper bound for a certain proportion of times. Confidence intervals measure the degree of uncertainty or certainty in a sampling method. Here we can say that we are sure that 95% of samples are falling between lower boud and upper bound, or we can say that we are 5% sure that our constructed interval will not contain data. Intervals are always symetric aound middle.

**In this problem, you're going to estimate the average of a random variable. In contrast with previous problems, this time the observations are collected one by one. To observe a single realization of this random variable, call `my_omicron(student_id)` function. Upon calling this function, a single observation/realization of this random variable is returned.**

**A) Draw 100 samples from my_omicron distribution. After observing each realization, calculate the sample mean of the realizations you've observed so far. In addition to that, construct a confidence interval for your estimation at 0.1 significane level. Finally, plot your estimations in a bar plot alongside their confidence errors as whisckers of each bar. Your final plot should be similar to this:**

<img src="https://support.content.office.net/en-us/media/68efe29e-f30c-4c6f-95cf-9cff7dde1131.jpg" width="400px" alt="no internet connection!">

```python
samples_myomicron_100 = np.array([])
mean_samples = np.array([])
confidentioal_interval = np.array([])
intervals = np.array([])
interval_co = 0.95
for i in range(100):
    samples_myomicron_100 =  np.append(samples_myomicron_100,my_omicron(810196514))
    sigma = np.sqrt(np.var(samples_myomicron_100, ddof = 1))/np.size(samples_myomicron_100)
    mean_samples = np.append(mean_samples, np.mean(samples_myomicron_100))
    if np.size(samples_myomicron_100) == 1:
        confidentioal_interval = np.append(confidentioal_interval, np.var(samples_myomicron_100, ddof = 0))/np.size(samples_myomicron_100)
    else:
        confidentioal_interval = np.append(confidentioal_interval, sigma)
    confidence_coefficient = stats.norm.ppf(interval_co)
    bound = confidence_coefficient*confidentioal_interval[i]
    intervals = np.append(intervals, np.array([bound, bound]))
transposed_intervals = intervals.reshape((100, 2)).transpose()   

fig = plt.figure(figsize=(45,30))
fig.patch.set_facecolor('yellow')
fig.patch.set_alpha(0.47)
axes = fig.add_subplot(111)
axes.errorbar(range(0,100), mean_samples, transposed_intervals, fmt = 'o-', capsize = 5,capthick = 2, color = 'black', ecolor = 'tomato')
axes.patch.set_facecolor('red')
axes.patch.set_alpha(0.2)
axes.yaxis.label.set_size(40)
axes.xaxis.label.set_size(40)
plt.title("Distribution of 100 samples mean with confidential intervals as whiskers plot from my_omicron", fontsize = 30)
plt.xlabel('Samples magnitude deviation', fontsize = 30)
plt.ylabel('Density of samples in each bin', fontsize = 30)
plt.show()
```
